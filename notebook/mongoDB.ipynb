```python
import pandas as pd
import pymongo
import os
from dotenv import load_dotenv
from tqdm.auto import tqdm
```

```python
mongo_user = os.getenv('MONGO_USER')
mongo_pass = os.getenv('MONGO_PASS')
mongo_cluster = os.getenv('MONGO_CLUSTER')
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
data = df.to_dict(orient='records')
```

## Pushing data to MongoDB

```python
DB_NAME = "mental_health_mlops"
COLLECTION_NAME = "raw_data"
CONNECTION_URL = f'mongodb+srv://{mongo_user}:{mongo_pass}@{mongo_cluster}/?retryWrites=true&w=majority&appName=Cluster0'
```

```python
## Project -> Database -> Collection (data)
client = pymongo.MongoClient(CONNECTION_URL)
data_base = client[DB_NAME]
collection = data_base[COLLECTION_NAME]
```

```python
## uploading data to MongoDB
## file size is too large.. uploading in batches
BATCH_SIZE = 500
for i in tqdm(range(0, len(data), BATCH_SIZE)):
    collection.insert_many(data[i:i+BATCH_SIZE])
```
